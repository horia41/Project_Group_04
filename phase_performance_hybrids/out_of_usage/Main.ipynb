### Hybrid: CSWinFPN

FPN with CSWin as backbone

```python

import torch
import torch.nn as nn
# import torch.nn.functional as F
# import torch.optim as optim
# from torch.utils.data import DataLoader
# import torchvision
# import torchvision.transforms as transforms
# from PIL.Image import Image
# import cv2
# from tqdm import tqdm

from phase_performance_hybrids.out_of_usage.cswin_fpn_hybrid import CSWinFPN
```

```python
# input must be (img_size, img_size) and divisible by 32
img_size = 256
cswin_kwargs = dict(
    img_size=img_size,
    in_chans=3,
    embed_dim=64,
    depth=[2, 2, 6, 2],
    split_size = [8, 8, 4, 2], # Must be: [64%x==0, 32%x==0, 16%x==0, 8%x==0]
    num_heads=[2, 2, 2, 2],
    mlp_ratio=4.0, 
    qkv_bias=True,
    # qk_scale=None,
    # drop_rate=0.0,
    # attn_drop_rate=0.0,
    # drop_path_rate=0.0,
    norm_layer=nn.LayerNorm,
    # use_chk=False,
)

model = CSWinFPN(
    cswin_kwargs=cswin_kwargs, 
    fpn_out_channels=128, 
)
# model = CSWinBackbone(**cswin_kwargs)

model.eval()
with torch.no_grad():
    x = torch.randn(1, 3, img_size, img_size)
    pyramids = model(x)
    for name, t in pyramids.items():
        print(f"{name}: shape={t.shape}")
```
